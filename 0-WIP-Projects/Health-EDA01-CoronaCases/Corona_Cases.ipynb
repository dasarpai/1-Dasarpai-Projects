```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("covid-19-all.csv")
df['China']=df["Country/Region"].apply(lambda x: 1 if x=="China" else 0)
```

```python
df
```

```python
plt.figure(figsize=(15,5))
df1=df.groupby(["China","Date"]) 
df1.sum()
df1=df1[["Confirmed","Recovered","Deaths"]].sum().reset_index()
df_china=df1[df1.China==1]
df_nochina=df1[df1.China==0]
sns.lineplot(x="Date", y="Deaths", data=df_china, label="China")
sns.lineplot(x="Date", y="Deaths", data=df_nochina, label="Rest")
plt.legend(ncol=2, loc="lower right", frameon=True)
plt.title("Corona Death Rate", fontsize=20)
plt.xticks(rotation=90, fontsize=9)
plt.show()
```
